```python
# Example: Standardize 'gender' and 'country' columns
if 'gender' in ab.columns:
    ab['gender'] = ab['gender'].astype(str).str.strip().str.lower().replace({
        'f': 'female', 'm': 'male', 'female': 'female', 'male': 'male'
    })

if 'country' in ab.columns:
    ab['country'] = ab['country'].astype(str).str.strip().str.title()
```

```python
# Automatically find and convert date columns
for col in ab.columns:
    if 'date' in col.lower():
        ab[col] = pd.to_datetime(ab[col], errors='coerce')
        # Format as dd-mm-yyyy
        ab[col] = ab[col].dt.strftime('%d-%m-%Y')
```

```python
# Clean column names: lowercase, no spaces or special characters
ab.columns = ab.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('[^a-z0-9_]', '', regex=True)
```

```python
# Convert 'age' column to integer
if 'age' in ab.columns:
    ab['age'] = pd.to_numeric(ab['age'], errors='coerce').fillna(0).astype(int)

# Reconvert any formatted dates back to datetime if needed
for col in ab.columns:
    if 'date' in col.lower():
        ab[col] = pd.to_datetime(ab[col], format='%d-%m-%Y', errors='coerce')
```
